Following few cells have python code written for log loss function, sigmoid AF, relu AF, leaky relu and tahh AF

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
def leaky_relu(x):
    return np.maximum(0.1*x,x)
```

Writing a python code for gradient descent for a neural network which has 4 inputs, single hidden layer with 3 neurons and an output neuron.
This function will be fed with all activation functions provided and after runing certain epochs, after adjusting its weights and biases and based upon the loss value of the model. Model, decides which activation function to be selected.

```python
def gradient_descent(variance,skewness,curtosis,entropy, y_true, epochs, af):
    
    for j in range(4):
        
        w1=w2=w3=w4=w5=w6=w7=w8=w9=w10=w11=w12=w13=w14=w15=1
        bias1=bias2=bias3=bias4=0
        avg_loss=[]
        y=[]

        rate=0.5
        n=len(x_train)

        for i in range(epochs):
            y1= [w1*variance + w2*skewness + w3*curtosis + w4*entropy + bias1,
                 w5*variance + w6*skewness + w7*curtosis + w8*entropy + bias2,
                 w9*variance + w10*skewness + w11*curtosis + w12*entropy + bias3,]


            o1,o2,o3= af[j](np.array(y1))

            y2 = w13*o1 + w14*o2 + w15*o3 + bias4
            op = af[j](np.array(y2))

            loss=round(log_loss(y_true,op),3)
            avg_loss.append(loss)
            
            w1d=w5d=w9d=(1/n)*np.dot(np.transpose(variance),(op-y_true))
            w2d=w6d=w10d=(1/n)*np.dot(np.transpose(skewness),(op-y_true))
            w3d=w7d=w11d=(1/n)*np.dot(np.transpose(curtosis),(op-y_true))
            w4d=w8d=w12d=(1/n)*np.dot(np.transpose(entropy),(op-y_true))

            w13d=(1/n)*y1[0]*(op-y_true)
            w14d=(1/n)*y1[1]*(op-y_true)
            w15d=(1/n)*y1[2]*(op-y_true)

            bias1_d=np.mean(op-y_true)
            bias2_d=np.mean(op-y_true)
            bias3_d=np.mean(op-y_true)
            bias4_d=np.mean(op-y_true)


            w1=w1-rate * w1d
            w2=w2-rate * w2d
            w3=w3-rate * w3d
            w4=w4-rate * w4d
            w5=w5-rate * w5d
            w6=w6-rate * w6d
            w7=w7-rate * w7d
            w8=w8-rate * w8d
            w9=w9-rate * w9d
            w10=w10-rate * w10d
            w11=w11-rate * w11d
            w12=w12-rate * w12d
            w13=w13-rate * w13d
            w14=w14-rate * w14d
            w15=w15-rate * w15d

            bias1=bias1-rate*bias1_d
            bias2=bias2-rate*bias2_d
            bias3=bias3-rate*bias3_d
            bias4=bias4-rate*bias4_d

        if mean(avg_loss) < 1: #Can  also use the minimum of all the losses, for this particular dataset this approach has been taken.
            
            print('Model is selecting {}'.format(af[j]),'with loss being {}'.format(mean(avg_loss)))
        print('')
    return 
```
